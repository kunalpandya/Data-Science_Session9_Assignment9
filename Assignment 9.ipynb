```python
import pandas as pd
import numpy as np
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
baby_names.info()
```

```python
# Delete unnamed columns

del baby_names['Unnamed: 0']
del baby_names['Id']
baby_names.head()
```

```python
# Show the distribution of male and female

np.round((baby_names['Gender'].value_counts())/len(baby_names)*100,2)
```

```python
# Show the top 5 most preferred names

names = baby_names.groupby("Name").sum()
names.head()
print (names.shape)
names.sort_values("Count", ascending = 0).head()
```

```python
# What is the median name occurence in the dataset

names[names.Count == names.Count.median()]
```

```python
# Distribution of male and female born count by states

baby_names.groupby(["State" , 'Gender'])['Count'].sum()
```
